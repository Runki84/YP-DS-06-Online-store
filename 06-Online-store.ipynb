# Проект: разработка рекомендательной модели для Интернет-магазина «В один клик»

## Фабула проекта

**Заказчик:** Интернет-магазин «В один клик» по продаже товаров различного ассортимента для разных сегментов покупателей от товаров для детей, мелкой бытовой техники до косметики продуктов питания.

**Данные:**

 - данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении;
 - данные о выручке, которую получает магазин с покупателя (сколько покупатель всего потратил за период взаимодействия с сайтом);
 - данные о времени (в минутах), которое покупатель провёл на сайте в течение периода;
 - данные о среднемесячной прибыли продавца за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю

**Задачи проекта:**

Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность:
 - промаркировать уровень финансовой активности постоянных покупателей;
 - собрать данные по клиентам по следующим группам:
    - признаки, которые описывают коммуникацию сотрудников компании с клиентом;
    - признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как часто;
    - признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине;
    - признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и сколько времени проводит на сайте;
 - построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца;
 - включить в исследование дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца;
 - используя данные модели и данные о прибыльности клиентов, выделить сегменты покупателей и разработать для них персонализированные предложения.

**План проекта:**

1. Загрузка данных:
    * Загрузить данные;
    * Проверить, что данные в таблицах соответствуют описанию.
    
    
2. Преобработка данных. Провести необходимую предобраблотку данных.


3. Исследовательский анализ данных:
    * Провести исследовательский анализ данных из каждой таблицы;
    * Отобрать клиентов с покупательской активностью не менее трёх месяцев, то есть таких, которые что-либо покупали в этот период;
    * Оформить выводы по результатам исследовательского анализа данных.
     
     
4. Объединение таблиц:
    * Объединить таблицы market_file.csv, market_money.csv, market_time.csv;   
    * Данные о выручке и времени на сайте находятся в одном столбце для всех периодов. В итоговой таблице сделать отдельный столбец для каждого периода.
     
     
5. Корелляционный анализ данных:
    * Провести корреляционный анализ признаков в количественной шкале в итоговой таблице для моделирования;
    * Сделать выводы о мультиколлинеарности и при необходимости устраните её;


6. Использование пайплайнов:
    * Во время подготовки данных использовать ColumnTransformer. Количественные и категориальные признаки обработать в пайплайне раздельно. Для кодирования категориальных признаков использовать как минимум два кодировщика, для масштабирования количественных — как минимум два скейлера;
    * Обучить четыре модели: KNeighborsClassifier(), DecisionTreeClassifier(), LogisticRegression() и  SVC(). Для каждой из них подобрать как минимум один гиперпараметр. Выбрать подходящую для задачи метрику, аргументировать выбор. Использовать эту метрику при подборе гиперпараметров;
    * Выбрать лучшую модель, используя заданную метрику. Применить одну из стратегий:
         - использовать пайплайны и инструменты подбора гиперпараметров для каждой модели отдельно, чтобы выбрать лучшую модель самостоятельно;
         - использовать один общий пайплайн для всех моделей и инструмент подбора гиперпараметров, который вернёт вам лучшую модель;


7. Анализ важности признаков;
    * Оценить важность признаков для лучшей модели и построить график важности с помощью метода SHAP;
    * Сделайть выводы о значимости признаков:
        - какие признаки мало значимы для модели;
        - какие признаки сильнее всего влияют на целевой признак;
        - как можно использовать эти наблюдения при моделировании и принятии бизнес-решений;
     
     
8. Сегментация покупателей;
    * Выполнить сегментацию покупателей. Использовать результаты моделирования и данные о прибыльности покупателей;
    * Выбрать группу покупателей и предложить, как увеличить её покупательскую активность:
        - провести графическое и аналитическое исследование группы покупателей;
        - сделать предложения по работе с сегментом для увеличения покупательской активности;
    * Сделайть выводы о сегментах:
        - какой сегмент взяли для дополнительного исследования;
        - какие предложения сделали и почему;
    
    
9. Общий вывод

### Паспорт данных

#### market_file.csv

Данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении:

 - **id** — номер покупателя в корпоративной базе данных.
 - **Покупательская активность** — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
 - **Тип сервиса** — уровень сервиса, например «премиум» и «стандарт».
 - **Разрешить сообщать** — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
 - **Маркет_актив_6_мес** — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
 - **Маркет_актив_тек_мес** — количество маркетинговых коммуникаций в текущем месяце.
 - **Длительность** — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
 - **Акционные_покупки** — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
 - **Популярная_категория** — самая популярная категория товаров у покупателя за последние 6 месяцев.
 - **Средний_просмотр_категорий_за_визит** — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.

## Использование пайплайнов

### Подготовка данных

Перед подготовкой пайплайнов, необходимо подготовить данные, для дальнейшей работы

```python
# подготовим данные для обучения
X = data_market.drop(['buy_activity'],axis=1)
y = data_market['buy_activity']

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify = y
)
```

```python
print('Размер выборок:')
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
print()

# проведем преобразование кодировщиком LabelEncoder()
le = LabelEncoder()
le.fit_transform(y_train.unique())
print ('Значениям покупательской активности', le.classes_, 'присвоены значения', le.transform(le.classes_))
y_train = le.transform(y_train)
y_valid = le.transform(y_valid)
```

Теперь поэтапно подготовим пайплан для обучения моделей

```python
# запишем переменные со списками колонок
ohe_columns = ['pop_cat']
ord_columns = ['service_type', 'message_offer']
num_columns = X.select_dtypes('number').columns
```

```python
# запишем пайплайн для подготовки признаков из списка ohe_columns
# используем заполнение пропусков и OHE-кодирование
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )
```

```python
# запишем пайплайн для подготовки признаков из списка ord_columns
# используем заполнение пропусков и Ordinal-кодирование
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'],
                    ['да', 'нет'],

                ],
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
# запишем общий пайплайн
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)
```

### Обучение моделей

На этом этапе соберем финальный вариант пайплайна который будет подготавливать данные и обучать разные модели.
Финальная версия пайплайна будет обучать модели:
 - KNeighborsClassifier(),
 - DecisionTreeClassifier(),
 - LogisticRegression(),
 - SVC()

Для сравнения результатов работы обученных моделей используем метрику ROC-AUC - метрика, которая отображаеть истинную и постоянную способность модели к прогнозированию. В идеале показатель ROC-AUC должен быть максимально близко к 1. Значение выбираем по валидационной выборке.

```python
# запришем итоговый пайплайн
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', KNeighborsClassifier())
])

param_grid = [
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 500),
        'models__max_features': range(1, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear',penalty='l1')],
        'models__C': range(1, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__kernel': ['poly', 'rbf', 'sigmoid'],
        'models__degree': range(1, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    }
]

model = RandomizedSearchCV(
                                pipe_final,
                                param_grid,
                                cv=5,
                                scoring='roc_auc',
                                random_state=RANDOM_STATE,
                                n_jobs=-1
).fit(X_train, y_train)


# выводим лучшую модель и ее параметры
print('Лучшая модель и её параметры:\n\n', model.best_estimator_)
# выводим лучшую метрику качества
print('\nМетрика roc_auc для лучшей модели:\n\n', model.best_score_)
# выводим значение площади ROC-кривой
print('\nПлощадь ROC-кривой на кросс-валидации:\n\n', model.best_score_)
```

```python
roc_auc_valid = roc_auc_score(y_valid, model.predict_proba(X_valid)[:, 1])
print('\nПлощадь ROC-кривой на валидационной выборке:\n\n', roc_auc_valid)
roc_auc_km = (roc_auc_valid - model.best_score_) / model.best_score_*100
```

```python
dummy_clf = DummyClassifier(strategy="stratified")
dummy_clf.fit(X_train, y_train)

y_pred_proba = dummy_clf.predict_proba(X_valid)[:, 1]
roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f"ROC AUC: {roc_auc}")
```

Для оценки важности признаков для лучшей модели были построены графики важности с помощью метода SHAP (значения Шепли). После пострения графиков уствновлено, что:
 - Наибольшее влияние оказывают следующие признаки в порядке убывания:
    - promo_buy - доля акционных покупок
    - mean_view_cat_visit - среднее количество просмотренных категорий
    - min_cur_month, min_prev_month - время в минутах проведенное на сайте
 - Наименьшее влияние оказывают следующие признаки в порядке возрастания:
    - популярная категория - кухонная посуда
    - тип сервиса
    - популярная категория - косметика и аксессуары
Однаков к важным признакам чтоит относиться с определенной осторожность, поскольку они являются взаимосвязанными и производными:
 - нельзя влиять на долю акционных покупок напрямую, для повышения активности покупателей. На долю акционных покупок, как отрицательно, так и положительно может влиять маркетинговая активность. Статистически этот показатель у нас в наименьших, как раз из-за того, что он может влиять как в положительную сторону, так и в отрицательную. Кроме того, на это могут влиять и ошибки сервиса.
 - среднее количество просмотенных категорий является косвенным показателем покупательской активности, однако причины этого признака могут быть и отрицательными - плохая навигация на сайте.
 - соответствено время проведенное на сайте хоть и может привести к покупательской активности, но его причиной не должна быть плохая навигация на сайте или ошибки. '

Тем не менее, исходя из полученных ранее данных, после устранения негативных факторов влияющих на признаки можно рекомендовать:
 - уделить внимание разработке рекомендательной системы на основании предпочтений пользователей, чтобы повысить конверсию посещения страниц сайта в покупки
 - разработать систему лояльности для удержания постоянных пользователей, без чрезмерного использования скидочной системы
 - при разработке рекомендательной системы и программы лояльности, учесть основной потрет потребителя - женщины с детьми.

<h3>Сегментация покупателей</h3>
По итогу работы по сегментации пользователей, все пользователи были разделены на два категории: высокоприбыльные и низкоприбыльные. С учетом ранее проведенной работы по обучени моделей, и нового категорирования, все пользователи были разделены на 3 сегмента:
- Высокоприбыльные активные
- Высокоприбыльные со снижающейся активностью
- Низкоприбыльные

В ходе изучения полученных сегментов, были выявлены следующие обстоятельства:
 - Пользователи, на текущий момент относящиеся к высокоприбыльным с высокой покупательской активностью, имеют тенденцию к снижению покупательской активности. Снижение отчасти вызвано техническими ошибками на сайте, отчасти чрезмерной маркетингоовой активностью. Проведение маркетинговых акций привело к повышению покупательской активности, однако эффект был кратковременным и общая выручка от этой категории покупателей начала снижаться.
 - Пользователи относящиеся к высокоприбыльным со снижающейся покупательской активностью, постепенно снижают объемы покупок и планомерно переходят в категорию малоприбыльных. При этом они не только снижают объемы покупок, но и отказываются от премиального сервиса и отказываются от получения маркетинговых сообщений.
 - Пользователи относящиеся к малоприбыльным показывают паразитическое поведение: покупают часто акционные товары, много времени проводят на сайте сранивая товары с конкурентами, покупают только дешевые товары. ъ

По результатам сегментирования пользователей были даны рекомендации по работке с каждыйм сегментом для повышения их прибыльности и активности, а именно:

1. Общие рекомендации по сайту:
 - провести технический аудит сайта - большое количество ошибок приводит к накоплению негативной реакции у пользователей. Кроме того, ошибки сайта могут служить причиной снижения покупательской активности напрямую из-за невозможности оплатить купленный товар
 - провести аудит навигации на сайте, слишком сложная навигация может приводить к повышению времени просмотра и количества страниц на сайте, но при этом снижению конверсии просмотров товарных позиций в покупки. Также слишком сложная навигация на сайте приводит к накоплению негативной реакции у пользователей.
 - уточнить детали по отличию сервисов Премиум и Стандарт, с учетом текущий вводных нет ясности влияния категории сервиса на маркетинговые активности.
2.  Общие рекомендации по маркетингу. Для повышения эффективности марктеингово взаимодействия с посетителями необходимо:
 - делать маркетинговые активности системными, в соответствии с заранее определенной стратегией и графиком, минимум на год, с учетом портрета пользователя,
 - систематизировать маркетинговые коммуникации до уровня, по возможности не более 1 раза в неделю, но и не менее раза в две недели, чтобы стимулировать активность пользователей.
 - сами маркетинговые активности организовывать с учетом их экономической эффективности и задачи стимулирования спроса. Как пример: сделать скидку на товар популярной категории, при покупки товара менее популярной категории.
 - не смотря на первичные вводны видим отток кользователей с сайта, в связи с этим снова запустить рекламную кампанию по привлечению на сайт новых пользователей, а запустить рекламную кампанию по пользователям с минимальной активностью на сайте, с целью их возврата в число активных пользователе;
 - внедрить для всех сегментов рекомендательную систему на основании предпочтений пользователей, чтобы повысить конверсию посещения страниц сайта в покупки.
3. Маркетинг по сегментам:
  - в сегменте высокоприбыльных и активных пользователей основная задача маркетинга должна состоять в удержании активности пользователей:
      - организовать маркетинговые активности с целью их удержания и прекращения перетока в менее прибыльные и менее активные категории. Примерами активностей могут быть розыгрыши призов именно в этой категории.
      - организовать программу лояльности для этого сегмента таким образом, чтобы они получали дополнительные бонусы при совершении определенного количества покупок в течение определенного периода времени (Пример: косметическое средство, при совершении покупок более 15000 рублей в квартал).
      - стимулировать покупку менее популярных товаров вместе с покупкой наиболее популярных (Пример: скидка на товары для детей при покупке косметики или домашнего текстиля).
 - в сегменте высокоприбыльных пользователей со снижающейся активностью основная задача маркетинга должна состоять в повышении их покупательской активности:
     - в течение первого месяца, организовать акции на наиболее популярные у них товары, чтобы стимулировать покупки.
     - в следуюшие месяцы стимулировать покупки на менее популярные категории, как в предыдущем сегменте
     - стимулировать оформление сервиса Премиум, чтобы он был им удобнее в использовании (Например выбор категорий на короые будут присылать маркетинговые сообщения, или выбор категорий на которые будет в течение месяца действтвать скидка, или выбор периодичности получения маркетиновоых сообщений)
     - за счет скидок на товары стимулировать оформление согласия на получени маркетинговых сообщений (напимер индивидуальными промокодами, при оформлении согласия).
 - в сегменте малоприбыльных, необходимо стимулировать их больше тратить именно на нашем сайте, чтобы повысить их прибыльность:
     - проводить акции на условиях покупки более одного товара
     - предоставлять им предложения из смежных товаров (Например рекомендовать, что при покупке одного товара часто покупают смежный)
     - сделать акцию, по предоставлени скидки на товар, если у конкурента он дешевле, чтобы остановить отток к конкурентам
     - исключительно в этом сегменте проводить акции по типу "черной пятницы".
     
Указанные меры позволят остановить отток пользователей, восстановить их покупательскую активность и увеличить уровень прибыльности.
